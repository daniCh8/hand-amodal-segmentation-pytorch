```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
i_names = [x for x in os.listdir('./data_sample/') if x.startswith('img')]
m_names = [x for x in os.listdir('./data_sample/') if x.startswith('msk')]
```

```python
imgs = [np.transpose(np.load('./data_sample/{}'.format(i)), (1,2,0)) for i in i_names]
msks = [np.load('./data_sample/{}'.format(i)) for i in m_names]
```

```python
def show_pair(index=-1):
    if index == -1:
        index = np.random.randint(len(imgs))
    
    plt.figure(figsize=(10,8))
    ax = plt.subplot2grid((1,3), (0,0))
    ax.imshow(imgs[index])
    ax.axis('off')
    
    ax = plt.subplot2grid((1,3), (0,1))
    ax.imshow(msks[index][0])
    ax.axis('off')
    
    ax = plt.subplot2grid((1,3), (0,2))
    ax.imshow(msks[index][1])
    ax.axis('off')
```

```python
show_pair()
```
